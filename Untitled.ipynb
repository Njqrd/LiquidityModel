```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
def generate_shift_flag(data):
    for contract in sorted(data.keys()):
        current_contract_df = data[contract]
        
        try:
            next_contract = sorted(data.keys())[sorted(data.keys()).index(contract) + 1]
            next_contract_df = data[next_contract]

            # Check the Open Interest shift and assign the flag.
            # Since contracts can overlap, you might want to adjust the logic for identifying the shift.
            # Here, we simply compare the last OpenInt of the current contract with the first OpenInt of the next contract.
            if next_contract_df['OpenInt'].iloc[0] > current_contract_df['OpenInt'].iloc[-1]:
                current_contract_df['Shift'] = 1
            else:
                current_contract_df['Shift'] = 0
            
            data[contract] = current_contract_df

        except IndexError:  # For the last contract, there's no next contract
            current_contract_df['Shift'] = 0
            data[contract] = current_contract_df
    
    return data
```

```python
data = generate_shift_flag(data)
```

```python
def preprocess_data(df, feature_col='OpenInt', label_col='Shift'):
    # Convert the DataFrame to a numpy array
    dataset = df[feature_col].values
    labels = df[label_col].values
    dataset = dataset.astype('float32')

    # Normalize the dataset
    scaler = MinMaxScaler(feature_range=(0, 1))
    dataset = scaler.fit_transform(dataset.reshape(-1, 1))

    return dataset, labels, scaler
```

```python
def create_dataset(dataset, labels, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(labels[i + look_back])
    return np.array(dataX), np.array(dataY)
```

```python
def create_model(look_back):
    model = Sequential()
    model.add(LSTM(50, input_shape=(1, look_back)))
    model.add(Dense(1, activation='sigmoid'))  # change here for binary classification
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])  # change here for binary classification
    return model
```

```python
def train_model(model, trainX, trainY, batch_size=1, epochs=20):
    model.fit(trainX, trainY, epochs=epochs, batch_size=batch_size, verbose=2)
    return model
```

```python
def evaluate_model(model, testX, testY):
    testPredict = model.predict(testX)
    testPredict = (testPredict > 0.5).astype(int)
    
    from sklearn.metrics import classification_report
    print(classification_report(testY, testPredict))
    
    return testPredict
```

```python
for contract, df in data.items():
    dataset, labels, scaler = preprocess_data(df)
    look_back = 1
    train_size = int(len(dataset) * 0.67)
    train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
    train_labels, test_labels = labels[0:train_size], labels[train_size:len(labels)]
    trainX, trainY = create_dataset(train, train_labels, look_back)
    testX, testY = create_dataset(test, test_labels, look_back)
    trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
    testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
    model = create_model(look_back)
    model = train_model(model, trainX, trainY, epochs=20)
    testPredict = evaluate_model(model, testX, testY)
```
